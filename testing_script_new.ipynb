```python
SELECT 
t1.*,t3.PICKUP_COMPLETED,t4.ORDER_DELIVERED
FROM
(select 
DISTINCT PRD_ORDER._ID as _ID,
PRD_ORDER.CREATEDAT,
PRD_ORDER.DELIVERYADDRESS_CITY,
PRD_ORDER.PICKUPADDRESS_ZIPCODE,
PRD_ORDER.PICKUPADDRESS_STATE,
PRD_ORDER.PICKUPADDRESS_CITY,
PRD_ORDER.DELIVERYADDRESS_STATE,
PRD_ORDER.DELIVERYADDRESS_ZIPCODE,
PRD_ORDER.TENANTID,
PRD_ORDER.PROVIDER,
PRD_ORDER.DELIVERYADDRESS_COUNTRY,
PRD_ORDER.PICKUPADDRESS_COUNTRY,
PRD_ORDER.DELIVERYADDRESS_REPORTLATITUDE,
PRD_ORDER.DELIVERYADDRESS_REPORTLONGITUDE,
PRD_ORDER.PICKUPADDRESS_LATITUDE,
PRD_ORDER.PICKUPADDRESS_LONGITUDE,
PRD_ORDER.HASPERISHABLEITEMS,
PRD_ORDER.HASREFRIGERATEDITEMS,
PRD_ORDER.ISBEERORWINE,
PRD_ORDER.ISFRAGILE,
PRD_ORDER.ISTOBACCO,
//PRD_ORDER.ESTIMATEDDELIVERYTIME,
PRD_ORDER.PICKUPTIMESTARTSAT,
PRD_ORDER.DROPOFFTIMESTARTSAT,
PRD_ORDER.DROPOFFTIMEENDSAT
 
from PRD_ORDER
where 
prd_order.createdat > '2022-10-1'
 and
prd_order.createdat < '2022-10-31' and
prd_order.type = 'delivery' and 
prd_order.status ='ORDER_DELIVERED' and 
prd_order.PICKUPTIMESTARTSAT is not null and
prd_order.DROPOFFTIMESTARTSAT is not null and
prd_order.DROPOFFTIMEENDSAT is not null and
ISDROPOFFASAP = FALSE and
ISPICKUPASAP = FALSE
)t1


left join
(select DISTINCT PRD_ORDER._ID as _ID,CONGNIDA_ORDER_STATUS.UPDATEDATEPOCH as PICKUP_COMPLETED
from CONGNIDA_ORDER_STATUS,PRD_ORDER 
where CONGNIDA_ORDER_STATUS.ORDERID = PRD_ORDER._ID and CONGNIDA_ORDER_STATUS.STATUS = 'PICKUP_COMPLETED'//ETA
//order by PRD_ORDER._ID,CONGNIDA_ORDER_STATUS.STATUS 
)t3
ON t1._ID = t3._ID

left join
(select DISTINCT PRD_ORDER._ID as _ID,CONGNIDA_ORDER_STATUS.UPDATEDATEPOCH as ORDER_DELIVERED
from CONGNIDA_ORDER_STATUS,PRD_ORDER 
where CONGNIDA_ORDER_STATUS.ORDERID = PRD_ORDER._ID and CONGNIDA_ORDER_STATUS.STATUS = 'ORDER_DELIVERED' //ETD
//order by PRD_ORDER._ID,CONGNIDA_ORDER_STATUS.STATUS 
)t4
ON t1._ID = t4._ID
where  
t3.PICKUP_COMPLETED IS NOT NULL
and t4.ORDER_DELIVERED IS NOT NULL;
```
